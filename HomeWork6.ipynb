Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1


```python
a = list(map(int, input().split()))
k = int(input())
if k in a:
    b = a.index(k)
else:
    b = -1
print(b)
```

W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
a = list(map(str, input().split()))
b = {}
for i,ai in enumerate(a):
    b.setdefault(ai,[]).append(i)
c = {k:tuple(v) for k,v in b.items()}
print(c)
```

W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1


```python
a = list(map(int, input().split()))
k = int(input())
b = []
for i in range(k+1):
    if i in a:
        b.append(a.index(i))
    else:
        b.append(-1)
print(b)
        
```

```python
a = list(map(int, input().split()))
k = int(input())
vitri = {val:i for i, val in enumerate(a) }
b = [vitri.get(i, -1) for i in range(k+1)]
print(b)
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
a = list(map(int, input().split()))
tang = all(x < y for x, y in zip(a, a[1:]))
giam = all(x > y for x,y in zip(a, a[1:]))
if tang or giam:
    print(True)
else:
    print(False)
```
