```python
class Shape:
    def area():
        pass
    
class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self): return self.w*self.h
    
class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14*self.r**2
    
classes = { 'RECTANGE': Rectangle, 'CIRCLE': Circle }
a = input().strip()
b = map(float, input().split())
obj = classes[a](*b)
print(f'{obj.area():.2f}')
```

```python
class xe:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def __str__(self):
        return f' { self.brand}-{self.speed}'

class oto(xe):
    def __init__(self, brand, speed, seats):
        super().__init__(brand, speed)
        self.seats = seats
    def __str__(self):
        chuoicha = super().__str__()
        return f'{chuoicha}-{self.seats}'
    
classes = { 'Vehicle':xe,'Car':oto }
loai = input().strip()
dulieu = input().split()
dulieu[1] = int(dulieu[1])
if len(dulieu)> 2:
    dulieu[2] = int(dulieu[2])
o = classes[loai](*dulieu)
print(o)
```

```python
class polygon:
    def perimeter(): pass
    
class triangle(polygon):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        cv = self.a + self.b + self.c 
        return cv
        
    def __str__(self):
        return f'{self.perimeter()}'

a = map(int, input().split())
o = triangle(*a)
print(o)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f'{self.name} {self.age}'
    
class Student(Person):
    def __init__(self, name, age, school):
        super().__init__(name, age)
        self.school = school
    def __str__(self):
        a = super().__str__()
        return f'{a} {self.school}'
    
classes = {'Person':Person, 'Student':Student}
loai = input().strip()
tt = input().split()
o = classes[loai](*tt)
print(o)
        
```

```python
class Account:
    def __init__(self, balance ):
        self.balance = balance
    def deposit(self, x):
        self.x = x
        self.balance += self.x
    def get_balance(self):
        print(self.balance)

class tietkiem(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate
    def interest(self):
        return self.balance*self.rate
    def add_interest(self):
        return self.balance + self.interest()
    
a = map(float, input().split())
o = tietkiem(*a)
print(f'{o.add_interest():.2f}')
```

```python
class Printer:
    def __init__(self, x):
        self.x = x
    def __str__(self):
        return f'{self.x}'

class BinaryPrinter(Printer):
    def __init__(self, x):
        super().__init__(x)
    def __str__(self):
        a = bin(self.x)
        b = a[2:]
        return f'{b}'
n = int(input())
o = BinaryPrinter(n)
print(o)
```

```python
class Matr:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def det(self):
        return self.a*self.d - self.b*self.c
    def __str__(self):
        return f'{self.trace()} {self.det()}'

a = map(int, input().split())
o = Matr(*a)
print(o)
```

```python
class product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class giam(product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent
    def ham(self):
        a = self.price*(100 - self.percent)/100
        return a
    def __str__(self):
        return f'{self.name} co gia {self.ham():.2f}'
    
a = input().split()
a[1] = int(a[1])
a[2] = int(a[2])
o = giam(*a)
print(o)
```
