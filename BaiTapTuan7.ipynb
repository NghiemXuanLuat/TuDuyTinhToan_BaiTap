Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target):


```python
def ham(ds, t, p, k):
    res = -1
    while t <= p:
        m = (t+p)//2
        if ds[m] == k:
            return m
        elif ds[m] < k:
            t = m+1
        else:
            p = m-1
    return res


ds = list(map(int, input().split()))
k = int(input())
print(ham(ds, 0, len(ds)-1, k))
```

W7A3
Viết hàm thực hiện thuật
toán bubble sort, Selection
Sort để sắp xếp danh sách
theo thứ tự tăng dần. Sắp
xếp nổi bọt (Bubble Sort);

```python
def ham(a):
    n = len(a)
    for i in range(n):
        doi = False
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                doi = True
        if not doi:
            break
    return a
a = list(map(int, input().split()))
print(ham(a))
# noi bot
```

```python
def ham(a):
    n = len(a)
    for i in range(n):
        nho = i
        for j in range(i+1, n):
            if a[j]< a[nho]:
                nho = j
                a[i], a[nho] = a[nho], a[i]
    return a
a = list(map(int, input().split()))
print(ham(a))
```

trong danh sách. Nếu có
nhiều phần tử cùng số lần,
trả về phần tử xuất hiện

```python
def count(k, a):
    b = 0
    for i in a:
        if i == k:
            b += 1
    return b


a = list(map(int, input().split()))    
ts = 0
gt = None
for i in a:
    if count(i,a) > ts:
        ts = count(i,a)
        gt = i
print(gt, ts)         
```

```python
def ham(a):
    b = [1]*len(a)
    for i in range(len(a)):
        for j in range(i):
            if a[i] > a[j]:
                b[i] = max(b[i], b[j]+1)
    return max(b)

a = eval(input())
print(ham(a))
                
```

W7A8
Bạn được cung cấp: một
danh sách các khoảng thời
gian interval = [[l1, r1], [l2,
r2], ...] và một danh sách
các truy vấn queries = [q1,
q2, ...]
Với mỗi truy vấn q, hãy tìm
độ dài nhỏ nhất của
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list

```python
import bisect
def ham(khoang, tv):
    q_sx = sorted([(gt, i) for i, gt in enumerate(tv)])
    khoang.sort(key = lambda x : x[1] - x[0] +1)
    kq = [-1] * len(tv)
    for l,r in khoang:
        dai = r - l +1
        vt = bisect.bisect_left(q_sx, (l, -1))
        while vt < len(q_sx) and q_sx [vt][0] <= r:
            gt, i_goc =  q_sx[vt]
            kq [i_goc] = dai 
            q_sx.pop(vt)
    return kq

a = eval(input())
b = eval(input())
print(ham(a,b))   
    
```

Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].

Efficient solution: using
heap or binary search tree

```python
a = eval(input())
b = [0]*len(a)
for i in range(len(a)):
    dem = 0
    for j in range(i+1, len(a)):
        if a[j] < a[i]:
            dem += 1
    b[i] = dem
print(b)
```

lớn nhất và giá trị nhỏ nhất

dãy hài hòa con dài nhất có
thể có trong mảng giá trị
nguyên \( arr \) có \( n \)
phần tử
Ví dụ:
Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con
dài nhất là [3,2,2,2,3]


```python
a = list(map(int, input().split()))
b = sorted(list(set(a)))
c = [0]*len(b)
for i in range(len(b)):
    dem = 0
    for j in range(len(a)):
        if a[j] == b[i] +1 or a[j] == b[i] :
            dem += 1
    c[i] = dem
print(max(c))
```
