```python
def ham(n):
    s = list(str(n))
    ket = []
    for i in range(len(s)):
        for j in range(i+1, len(s)):
            tam = s[:]
            tam[i], tam[j] = tam[j], tam[i]
            if tam[0] != '0':
                somoi = int(''.join(tam))
                ket.append(somoi)
    if len(ket) > 0:
        return min(ket)
    else:
        return n
n = int(input())
print(ham(n))
    
```

```python
def ham(a,k):
    vuon = [0] + a + [0]
    for i in range(1,len(vuon)-1):
        if vuon[i-1] == vuon[i] == vuon[i+1] == 0:
            vuon[i] == 1
            k -= 1
        if k == 0:
            return True
    return False

a = list(map(int, input().split()))
k = int(input())
print(ham(a,k))
```

```python
def ham(a):
    x,y = 0,0
    dadi = {(0,0)}
    huong = {'U':(0,1), 'D':(0, -1), 'R':(1, 0), 'L':(-1, 0)}
    for i in a:
        dx, dy = huong[i]
        x += dx
        y += dy
        if (x,y ) in dadi:
            return True
        dadi.add((x,y))
    return False
a = input()
print(ham(a))
```
