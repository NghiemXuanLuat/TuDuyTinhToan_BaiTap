Nhập vào hai số nguyên a, b. Thực hiện in ra kết quả phép chia a/b. Bắt lỗi ngoại lệ khi b

```python
def ham(a):
    for i in range(len(a)):
        for j in range(i+1, len(a)):
            if a[i] > a[j]:
                a[i], a[j] = a[j], a[i]
    return a
a = list(map(int, input().split()))
k = int(input())
a = ham(a)[::-1]
b = a[:k]
print(*b)
```

```python
class cha:
    def __init__(self, chu, tk):
        self.chu = chu
        self.tk = tk
        self.du = 0
    def ham(self, n):
        if n > 0:
            self.du += n
        elif n == 0:
            print(f'{self.du:.2f}')
class con(cha):
    def ham(self, n):
        if n < 0:
            self.du += self.du*(-n/100)
        else:
            super().ham(n)
loai = input().strip()
chu = input().strip()
tk = input().strip()
a = int(input())

acc = con(chu, tk) if loai == 'Tai' else cha (chu, tk)
for _ in range(a):
    n = float(input())
    acc.ham(n)
    
```
