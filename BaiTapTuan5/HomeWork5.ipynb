Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x


```python
def ham(n):
    a = len(str(n))
    b = str(n)
    so = 0
    for i in b:
        so += (int(i))**a
    if so == n:
        return True
    return False

n = int(input())
print(ham(n))
        
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def ham(x1, y1, x2, y2, x3, y3):
    if x3 == x1 and y3 == y1:
        return True
    elif (x3 == x1 and y3 != y1) or (x3 != x1 and y3 == y1):
        return False 
    elif (x2 - x1)/(x3 - x1) == (y2 - y1)/(y3 - y1):
        return True
    else:
        return False
    
x1,y1,x2,y2,x3,y3 = map(float,input().split())
print(ham(x1, y1, x2, y2, x3, y3))
    
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).



```python
import math 
def ham(x1,y1,x2,y2,x3,y3):
    cos = ((x1-x2)*(x3-x2) + (y1 - y2)*(y3 - y2))/((((x1-x2)**2 + (y1 - y2)**2)**0.5)*((x3 - x2)**2 + (y3 - y2)**2)**0.5)
    goc = math.acos(cos)
    return math.degrees(goc)

x1,y1,x2,y2,x3,y3 = map(float, input().split())
print(ham(x1,y1,x2,y2,x3,y3)) 
```

W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0

```python
# khong dung duoc so 0
def ham(n):
    a = [0]*10
    while n != 0:
        a[n % 10] += 1
        if a[n % 10] == 2:
            return False
        n //= 10
    return True

n = int(input())
tong = 0
for i in range(1, n+1):
    if ham(i):
        tong += i
        
print(tong)
```
