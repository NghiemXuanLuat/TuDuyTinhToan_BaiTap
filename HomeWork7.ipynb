Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.


```python
def ham(a):
    for i in a:
        if a.count(i) != i:
            return False
    return True

def xoa(a):
    b = []
    for i in a:
        if a.count(i) == i:
            b.append(i)
        elif a.count(i) > i and b.count(i) < i:
            b.append(i)
            
        
    return b



a = list(map(int, input().split()))
if ham(a):
    print(ham(a))
else:
    print(xoa(a))
```

Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng
dần, lẻ giảm dần; ;

```python
a = list(map(int, input().split()))
chan = [i for i in a if i % 2 == 0]
chan = sorted(chan)
le = [i for i in a if i % 2 != 0]
le = sorted(le, reverse=True)
ket = chan + le
print(ket)
```

Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài
>=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí
tự cái.

```python
def ham(a):
    if a.isalpha() and a[0].isupper():
        for i in range(1, len(a)):
            if a[i].isupper():
                return False
        return True

a = input().split()
dem =0
for i in a:
    if ham(i):
        dem += 1
print(dem)
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau


```python
a = list(map(int, input().split()))
b = [0]*len(a)
bd = 0
for i in range(1, len(a)):
    if a[i] > a[i-1]:
        b[i] = b[i-1] +1
    elif a[i] <= a[i-1]:
        b[i] = 0
c = b.index(max(b))
for i in range(c-max(b), c+1):
    print(a[i], end= ' ')
```

vào từ bàn phím
