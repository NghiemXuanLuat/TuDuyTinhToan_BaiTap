```python
class tru:
    def __init__(self, bk, cao):
        self.r = bk
        self.h = cao
    def dtich(self):
        S = 2*(self.r*self.r)*3.14 + 2*3.14*self.r*self.h
        return S
    def ttich(self):
        V = 3.14*(self.r**2)*self.h
        return V
    
a = tru(3, 5)
print(a.dtich())
print(a.ttich())
```

```python
class toado:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def vitri(self):
        a = ''
        if self.x == self.y == 0:
            a = 'Goc toa do'
        elif self.x == 0:
            a = 'tren truc tung'
        elif self.y == 0:
            a = ' tren truc hoanh'
        else:
            a = ' khong nam tren truc'
        return a
    def kc(self):
        a = ((self.x)**2 + (self.y)**2)**0.5
        return a
a, b = map(int, input().split())
m = toado(a,b)
print(m.vitri())
print(m.kc())
```
